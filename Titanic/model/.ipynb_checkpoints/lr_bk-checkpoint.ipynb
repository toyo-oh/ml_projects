##### 3.Cleanup and Transform the Data(feature engineering) #####

```python
#set missing age method-1

from sklearn.ensemble import RandomForestRegressor

def set_missing_age(df):
    age_df = df[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
    #  df.valuesã€€â†’ Return a Numpy representation of the DataFrame.(version 0.23: to_numpy() is recommended)
    #  print(age_df.values)
    # df.as_matrix()â†’ å°†frameè½¬æ¢ä¸ºå®ƒçš„æ•°å­—æ•°ç»„è¡¨ç¤ºå½¢å¼(version 0.23.0: Use DataFrame.values() instead.)
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()
    # yå³ç›®æ ‡å¹´é¾„ï¼ˆç¬¬1åˆ— ageï¼‰
    y = known_age[:, 0]
    # Xå³ç‰¹å¾å±žæ€§å€¼ï¼ˆç¬¬2åˆ—å¼€å§‹çš„å…¶ä»–åˆ—ï¼‰
    X = known_age[:, 1:]
    
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)
    predictedAges = rfr.predict(unknown_age[:, 1::])
    
    # ç”¨å¾—åˆ°çš„é¢„æµ‹ç»“æžœå¡«è¡¥åŽŸç¼ºå¤±æ•°æ®
    df.loc[ (df.Age.isnull()), 'Age' ] = predictedAges 
    
    return df,rfr
```

```python
#set missing age method-2

# æ ¹æ®å¤šåˆ—çš„å€¼åšåˆ¤æ–­ï¼Œç”Ÿæˆæ–°çš„åˆ—å€¼
def set_sex_flg(name,sex):
    if 'Mrs' in name:
        sex_flg = 'Mrs'
    elif 'Miss' in name:
        sex_flg = 'Miss'
    elif 'Mr' in name:
        sex_flg = 'Mr' 
    else:
        if sex=='male':
            sex_flg='Mr'
        else:
            sex_flg = 'Miss'
    return sex_flg

def set_missing_age2(df):
    df_mean = df.groupby(df.sex_flg)['Age'].mean()
    print(df_mean)
    df_sex_flg = df.loc[(df.Age.isnull()), 'sex_flg'] 
    df.loc[(df.Age.isnull()), 'Age']  = df_sex_flg.apply(lambda x : df_mean.loc[x])
    
    return df
```

```python
#df_train,rfr= set_missing_age(df_train)
df_train['sex_flg']=df_train.apply(lambda x : set_sex_flg(x.Name, x.Sex), axis=1)
df_train = set_missing_age2(df_train)
```

```python
#set missing cabin method-1

def set_cabin_type(df):
    #æ‰¾å‡ºnotnullæˆ–isnullçš„è¡Œï¼Œåœ¨cabinåˆ—ä¸Šè®¾å€¼yesæˆ–no
    df.loc[(df['Cabin'].notnull()), 'Cabin'] = "Yes"
    df.loc[(df['Cabin'].isnull()), 'Cabin'] = "No"
    return df
```

```python
#set missing cabin method-2

def get_cabin_no(s):
    cabinno = s[1:]
    if ' ' in cabinno:
        cabinno =  cabinno.split(sep=' ')[0]
    if ' ' in cabinno or cabinno=='':
            cabinno = 0
    return int(cabinno)

def set_cabin_type2(df):
    df['Cabin_1']=df.loc[df['Cabin'].notna(),'Cabin'].apply(lambda x: x[0:1])
    df['Cabin_2']=df.loc[df['Cabin'].notna(),'Cabin'].apply(lambda x: get_cabin_no(x))
    df.loc[(df['Cabin_2'].isnull()),'Cabin_2']=0
    return df
```

```python
df_train = set_cabin_type2(df_train)
df_train.drop('Cabin',axis=1,inplace=True)
```

```python
#add child feature
def add_child(age):
    child_flg = 0
    if age <=12:
        child_flg= 1
    return child_flg
```

```python
df_train['Child'] = df_train['Age'].apply(lambda x: add_child(x))
```

```python
#add mother feature
def add_mother(sex_flg, parch):
    mother_flg = 0
    if sex_flg=='Mrs' and parch>1:
        mother_flg =1
    return mother_flg
```

```python
df_train['Mother'] = df_train.apply(lambda x: add_mother(x.sex_flg,x.Parch), axis=1)
```

```python
dummy_Cabin = pd.get_dummies(df_train.Cabin_1,  prefix="Cabin_1")
dummy_Embarked = pd.get_dummies(df_train['Embarked'],  prefix="Embarked")
dummy_Sex = pd.get_dummies(df_train['Sex'], prefix="Sex")
dummy_Pclass =pd.get_dummies(df_train['Pclass'], prefix="Pclass")
```

```python
df_new= pd.concat([df_train, dummy_Cabin,dummy_Embarked,dummy_Sex,dummy_Pclass], axis=1)
df_new.shape
```

```python
import sklearn.preprocessing as preprocessing
scaler = preprocessing.StandardScaler()

df_new['Age_scaled']= scaler.fit(df_new[['Age']]) .transform(df_new[['Age']])
df_new['Fare_scaled'] = scaler.fit(df_new[['Fare']]).transform(df_new[['Fare']])
df_new['Cabin_2_scaled'] = scaler.fit(df_new[['Cabin_2']]).transform(df_new[['Cabin_2']])

df_new.drop(['Cabin_1_T'], axis=1, inplace=True)
```
