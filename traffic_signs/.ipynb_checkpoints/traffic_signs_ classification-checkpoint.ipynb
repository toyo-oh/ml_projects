### 2.数据预处理 ###

```python
#①、Xデータ：RGB→YCrCb(Y=0.299 R + 0.587 G + 0.114 B)　
#②、Xデータ：0-1 scale
#③、Xデータ：直方图均衡化（图像增强）　https://www.jianshu.com/p/99e8ff2ff25d
#④、Yデータ：one-hot encoding　

def process_image_data(X,y):
    #Y = 0.299 R + 0.587 G + 0.114 B
    X[:,:,:,0] = 0.299 * X[:,:,:,0] + 0.587*X[:,:,:,1] + 0.114*X[:,:,:,2]
    
    #0-1 scale
    X = X /255.0
    
    #直方图均衡化（图像增强）
    for i in range(X.shape[0]):
        X[i] = exposure.equalize_adapthist(X[i])
    
    #one-hot encoding
    #np.eys→对角单位矩阵（Return a 2-D array with ones on the diagonal and zeros elsewhere.）
    #y = np.eye(n_classes)[y]

    return X,y

# Test for Cell
tmp_x = train_x[90:94,:,:,:]
tmp_y = train_y[90:94]
print(tmp_x.shape)
# print(tst)
aft_x,aft_y = process_image_data(tmp_x,tmp_y)
print(aft_x.shape)

plt.figure(figsize=(3,3))
plt.subplot(1,2,1)
plt.imshow(tmp_x[0],cmap='Greys_r')
plt.subplot(1,2,2)
plt.imshow(aft_x[0],cmap='Greys_r')
plt.show()
```

### 3.数据扩充 ###

```python
#图片旋转，图片旋转+拉伸，平移

#a.图像的平移
#沿着x方向tx距离，y方向ty距离，那么需要构造移动矩阵M=[[1,0,tx],[0,1,ty]]：
#通过numpy来产生这个矩阵，并将其赋值给仿射函数cv2.warpAffine().
#仿射函数cv2.warpAffine()接受三个参数，需要变换的原始图像，移动矩阵M以及变换的图像大小

#b.图像的旋转
#图像的旋转矩阵一般为： M=[[cos(θ),-sin(θ)],[sin(θ),cos(θ)]]
#但是单纯的这个矩阵是在原点处进行变换的，为了能够在任意位置进行旋转变换，opencv采用了另一种方式：M=[[],[]]
#为了构造这个矩阵，opencv提供了一个函数：
#cv2.getRotationMatrix2D()，这个函数需要三个参数，旋转中心，旋转角度，旋转后图像的缩放比例]

#c.图像的彷射
#图像的旋转加上拉升就是图像仿射变换，仿射变化也是需要一个M矩阵就可以，但是由于仿射变换比较复杂，一般直接找很难找到这个矩阵，
#opencv提供了根据变换前后三个点的对应关系来自动求解M。这个函数是：
#M=cv2.getAffineTransform(pos1,pos2),其中两个位置就是变换前后的对应位置关系。输出的就是仿射矩阵M。

def transfor_image(img, mov_range, rot_range, trans_range):
    rows, cols, ch = img.shape
    
    # move
    tx = mov_range*np.random.uniform()-mov_range/2
    ty = mov_range*np.random.uniform()-mov_range/2
    Mov_M = np.float32([[1,0,tx],[0,1,ty]])
    img = cv2.warpAffine(img, Mov_M, (rows, cols))

    # rotation
    rot_range = np.random.uniform(rot_range)-rot_range/2
    Rot_M = cv2.getRotationMatrix2D((cols/2,rows/2),rot_range,1)
    img = cv2.warpAffine(img, Rot_M, (rows,cols))

    # trans
    pts1 = np.float32([[5, 5], [20, 5], [5, 20]])
    pt1 = 5+trans_range*np.random.uniform()-trans_range/2
    pt2 = 20+trans_range*np.random.uniform()-trans_range/2
    pts2 = np.float32([[pt1, 5], [pt2, pt1], [5, pt2]])
    trans_M = cv2.getAffineTransform(pts1, pts2)
    img = cv2.warpAffine(img, trans_M, (rows,cols))
    
    return img
```

```python
#不足的类目增加数量
# sign_classes, class_counts 
def create_left_counts(class_counts):
    max_class = np.max(class_counts)
    left_counts = []
    for count in class_counts:
        left_counts.append(max_class - count)
    return left_counts

left_counts = create_left_counts(class_counts)

print(len(left_counts))
print(left_counts)
```

```python
#获取某个特定分类的图片集
def get_same_class(X,y,class_i):
    same_class_X = np.ndarray(shape=(0,X.shape[1],X.shape[2],X.shape[3]), dtype=np.float32)
    same_class_y = np.ndarray(shape=(0), dtype=np.int32)
    
    for i in range(y.shape[0]):
        if y[i] == class_i:
            same_class_X = np.append(same_class_X, [X[i,:,:,:]], axis=0)
            same_class_y = np.append(same_class_y, [y[i]], axis=0)
    return same_class_X,same_class_y

# Test for Cell
tmp_x,tmp_y = get_same_class(train_x,train_y,sign_classes[0])
print(tmp_x.shape)
```

```python
#扩充某个特定分类图片集
def create_genre_images(X,y,class_i):
    #需要扩充的数量
    base_count = class_counts[class_i]
    left_count = left_counts[class_i]
    
    #重复次数
    add_unit = left_count // base_count
    add_single = left_count % base_count
    #原来该分类的数据集
    base_genre_X, base_genre_y = get_same_class(X, y, class_i)
    
    #整数倍的以unit单位扩充
    for i in range(add_unit):
        add_unit_X = np.ndarray(shape=(0,X.shape[1],X.shape[2],X.shape[3]), dtype=np.float32)
        add_unit_y = np.ndarray(shape=(0), dtype=np.int32)
        for idx in range(base_count):
            trans_img = transfor_image(base_genre_X[idx], 20, 10, 5)
            add_unit_X = np.append(add_unit_X, [trans_img], axis=0)
            add_unit_y = np.append(add_unit_y, [class_i], axis=0)
        base_genre_X = np.append(base_genre_X, add_unit_X, axis=0)
        base_genre_y = np.append(base_genre_y, add_unit_y, axis=0) 
        
    #余数的每张为单位扩充
    for j in range(add_single):
        rnd_img = transfor_image(base_genre_X[j], 20, 10, 5)
        base_genre_X = np.append(base_genre_X, [rnd_img], axis=0)
        base_genre_y = np.append(base_genre_y, [class_i], axis=0)
        
    if base_genre_X.shape[0] > 1000:
        base_genre_X = base_genre_X[0:1000,:, :, :]
        base_genre_y = base_genre_y[0:1000]
        
    print(base_genre_X.shape)
    print(base_genre_y.shape)
    return base_genre_X,base_genre_y

# Test for Cell
add_0_X,add_0_y = create_genre_images(train_x,train_y,sign_classes[0])
```

### 4.建立模型 ###

```python
#对每一类进行扩充数据
add_genre_X = np.ndarray(shape=(0,train_x.shape[1],train_x.shape[2],train_x.shape[3]), dtype=np.float32)
add_genre_y = np.ndarray(shape=(0), dtype=np.int32)

for j in sign_classes:
    add_x,add_y = create_genre_images(train_x, train_y,j) 
    add_genre_X = np.append(add_genre_X, add_x, axis=0)
    add_genre_y = np.append(add_genre_y, add_y, axis=0)

print(add_genre_X.shape)
print(add_genre_y.shape)
print(type(add_genre_y[0]))
```

```python
#对每张图片做处理
add_genre_X, add_genre_y = process_image_data(add_genre_X,add_genre_y)
```

### 5.模型参数编码 ###

```python
test_x, test_y = process_image_data(test_x,test_y)
```

```python
print(test_x.shape)
print(test_y.shape)
```

```python
IMG_HEIGHT = add_genre_X.shape[1]
IMG_WIDTH = add_genre_X.shape[2]

model = Sequential([
    Conv2D(32, (5, 5), kernel_regularizer = regularizers.l2(0.001),activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 1)),
    Dropout(0.3),
    MaxPooling2D((2, 2)),
    Conv2D(64, (5, 5), kernel_regularizer = regularizers.l2(0.001),padding='same', activation='relu'),
    Dropout(0.3),
    MaxPooling2D((2, 2)),
    Conv2D(128, (5, 5), kernel_regularizer = regularizers.l2(0.001),padding='same', activation='relu'),
    Dropout(0.3),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(1024, activation='relu'),
    Dropout(0.3),
    Dense(43, activation='softmax')
])
```

```python
model.summary()
```

```python
STEPS_PER_EPOCH = 1
lr_schedule = InverseTimeDecay(0.001,decay_steps=STEPS_PER_EPOCH*1000,decay_rate=0.95,staircase=False)
model.compile(optimizer=tf.keras.optimizers.Adam(lr_schedule),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
# This callback will stop the training when there is no improvement in
# the validation loss for three consecutive epochs.
# model.fit(data, labels, epochs=100, callbacks=[callback],
#     validation_data=(val_data, val_labels))

history = model.fit(add_genre_X[:,:,:,0:1], add_genre_y, epochs=10, callbacks=[callback],
                    validation_data=(test_x[:,:,:,0:1], test_y))
```
